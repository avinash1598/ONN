```python
"""
THETA: Learnt theta
THETA': Test signal theta

THETA - THETA' = omega2(2*m1 -+ n)*pi - omega1*(2*m2)*pi
Difference should be integer multiple of pi

Above is incorrect!!!!

(omega0*t + phi0) - phi is integer multiple of 2*pi and not
phi0 - phi.
"""
diff = ( o.bin_conn*(o.THETA - o.THETA_Test) )%np.pi
print(diff)

#Can't find phases of individual frequency 
# component as eqation were simulated using oscillator dynamical equation.
# Only way to find it would be to use spectral frequency graphs.
# Below calculation is valid for both small and large values of Aij
# Should be close to either 0 or 6.28. However, can't use this for non-stationary EEG signal.
curr_time = o.DT*o.NSAMPLES
ss_test_phi = curr_time*o.OMEGA_Test + o.PHI_Test
diff_phi = (o.PHI - ss_test_phi)%(2*np.pi)
print(diff_phi)

# f = plt.figure()
# for i in range(o.N):
#     for j in range(o.N):
#         #Either zero or close to pi
#         plt.axhline(y=diff%(np.pi), color='b', linestyle='--', label="diff")
# plt.ylabel("THETA")
# #plt.legend()
# plt.show()
plt.close('all')

#Show that the relationship can be satisfied for unique value of theta given that omega is given.
```
